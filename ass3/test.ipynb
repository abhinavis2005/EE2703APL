## Plotting d3

```python
x = []
y = []
with open("d3.txt", "r") as file:
    data = file.readlines()
    for i in data:
        x.append(float(i.split(",")[0]))
        y.append(float(i.split(",")[1].strip()))

x = np.array(x)
# x *= np.power(10,6)
y = np.array(y)
x_cut = x[50:]
y_cut = y[50:]
plt.plot(x_cut, y_cut)
```

## generating the fucntion

```python
def planksLaw(Ld, T,h,c, k ):
    B = (2*h*np.power(c,2))/(np.power(Ld,5)*(np.exp(h*c/(Ld*k*T))-1))
    return B
```

```python

p0 = np.array([1,1e-9,3e8,1])
params, params_covariance = sp.optimize.curve_fit(planksLaw, x_cut, y_cut, p0 = p0)
```

```python
print(params)
plt.figure(); plt.plot(x, y); plt.plot(x[50:], planksLaw(x[50:], *params))
# plt.plot(x_cut, y_cut)
```

```python
print(params)
```

### let us fix every other variable


```python
initparams = np.array([6.64*1e-34, 3*1e8,1.38e-23 ])

def newFun(Ld, T):
    return planksLaw(Ld,T, *initparams)

params, params_covariance = sp.optimize.curve_fit(newFun, x, y, p0 = [4000])
plt.plot(x,y)
plt.plot(x, newFun(x,*params))
print(params)
```
